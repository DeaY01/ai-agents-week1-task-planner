```python
!pip install openai -q

import openai
import os

# Optional: Paste your OpenAI key here for 10× smarter output (leave blank = still works)
openai.api_key = ""   # ← put your key here or leave empty

def task_planner(goal):
    prompt = f"""You are TaskPlanner Agent v1.0 — ruthless and autonomous.
Goal: {goal}

Break it into 5–12 tiny, sequential, idiot-proof steps.
Next to each step write the exact tool/API needed in *italics*.
Mark parallel steps with →
End with: Ready for autonomous execution? Yes/No
Be brutally specific. No fluff."""

    if openai.api_key and openai.api_key.strip() != "":
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )
        return response.choices[0].message.content.strip()
    else:
        return f"No API key → simple fallback plan:\n1. Research {goal}\n2. List steps\n3. Execute manually\n(Add your OpenAI key above for full power)"

# ←←← EDIT THIS LINE WITH YOUR GOAL AND RUN THE CELL ↓↓↓
your_goal = "Plan a weekend trip from Lagos State to Abeokuta under ₦150,000 for two people"

print("\nTaskPlanner Output:\n")
print(task_planner(your_goal))
```
